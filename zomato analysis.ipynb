```python
import pandas as pd
import numpy as np
import glob
```

```python
path = 'C:/Users/kajad/Downloads/ineuron/Data_Sets/zomato/zomato/Zomato/Bengaluru'  #getting the path of the csv files in local system
```

```python
csv_files_list = glob.glob(path+"/*")   #accessing all the csv files which returns in the form of list
csv_files_list
```

```python
#creating empty dataframe

csv_merg = pd.DataFrame()  
```

```python
#using for loop  to get the each file located in the local system
#concating each csv file to empty dataframe

for i in csv_files_list:
        csv_merg = pd.concat([csv_merg,pd.read_csv(i,sep='|')])   

csv_merg.head()
```

```python
#dropping unwanted lables in thed ataframe


zomato = csv_merg.drop(['URL','PAGE NO',],axis=1)
zomato
```

```python
#having a look at null vales in a column

zomato.loc[zomato['TIMING'].isnull()]
```

```python
#dropping those null values after having a look 

zomato.dropna(axis=1)
```

```python
#removing unwanted data in a column and resetting the indexes 
#this reset_index is used to drop the cthe current index column and replace it with increasing index integers


zomato = zomato.loc[zomato['RATING'] != 'NEW']
zomato = zomato.loc[zomato['RATING'] != 'Opening']
zomato = zomato.loc[zomato['RATING'] != '-'].reset_index(drop=True)
```

```python
#changing the data type of the labels for later usage 


zomato['RATING'] = zomato['RATING'].astype(float)
zomato['VOTES'] = zomato['VOTES'].astype(float)
```
